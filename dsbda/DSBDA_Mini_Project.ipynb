As there are many NULL values present in the given dataset. We need to replace those values by mean(in case of numerical data) or mode(in case of categorical data). Here, we need to work on "First Dose Administered" and "Second Dose Administered". Both of them are float, hence we will replace the Nan Values by mean(average).

#### For First Dose Administered

```python
# Average of First Dose Administered
avg_firstdose = data["First Dose Administered"].astype("float").mean(axis = 0)
print("Average of First Dose:", avg_firstdose)
```

```python
# Replacing First Dose Administered
data["First Dose Administered"].fillna(value = avg_firstdose, inplace=True)
data
```

#### For Second Dose Administered

```python
# Average of Second Dose Administered
avg_seconddose = data["Second Dose Administered"].astype("float").mean(axis = 0)
print("Average of Second Dose:", avg_seconddose)
```

```python
# Replacing Second Dose Administered
data["Second Dose Administered"].fillna(value = avg_seconddose, inplace = True)
data
```

**PIE CHART**

```python
# Calculating the total number of males vaccinated
male = data["Male(Individuals Vaccinated)"].sum()

# Calculating the total number of individuals vaccinated (for normalization)
total_vaccinated = data["Total Individuals Vaccinated"].sum()

# Calculating the percentage of males vaccinated
male_percentage = (male / total_vaccinated) * 100

# Calculating the percentage of females vaccinated
female_percentage = 100 - male_percentage

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie([male_percentage, female_percentage], labels=['Male', 'Female'], autopct='%1.1f%%')
plt.title('Distribution of Vaccinated Individuals by Gender')
plt.show()
```

**BAR CHART**

```python
# Grouping data by state and summing up the second dose administered
second_dose = data.groupby('State')['Second Dose Administered'].sum()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
second_dose.plot(kind='bar')
plt.title('Total Second Doses Administered by State')
plt.xlabel('State')
plt.ylabel('Total Second Doses Administered')
plt.xticks(rotation=90)
plt.show()
```

**BOXPLOT CHART**

```python
# Plotting a boxplot to visualize the distribution of total doses administered across different states
plt.figure(figsize=(12, 6))
sns.boxplot(x='State', y='Total Doses Administered', data=data)
plt.title('Distribution of Total Doses Administered by State')
plt.xlabel('State')
plt.ylabel('Total Doses Administered')
plt.xticks(rotation=90)
plt.show()
```
